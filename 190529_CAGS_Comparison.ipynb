```python
import gzip, json   # JSON is a very common way to store data objects
import pandas as pd # Very useful module for interacting with tables
import os

def read_cags_from_json(fp):
    # Make sure that the file exists
    assert os.path.exists(fp), "File doesn't exist %s" % (fp)
    
    # Read in the data object
    dat = json.load(gzip.open(fp, "rt"))
    
    return pd.DataFrame(
        [
            {
                "cag_id": cag_id,
                "gene_id": gene_id
            }
            for cag_id, gene_id_list in dat.items() # Iterate over keys and values
            for gene_id in gene_id_list
        ]
    )

read_cags_from_json("50_samples_sum_0.3_2019-01-09.cags.json.gz").head()
```
